### Задание №2 (Трекер бюджета в консоли)

```python
from task_2.transaction import Transaction
from task_2.budget_tracker import BudgetTracker
from task_2.transaction_type import TransactionType
from task_2.category import Category
```

```python
cats = [
    Category('школа', 500),
    Category('работа'),
    Category('отдых')
]
```

```python
budge_tracker: BudgetTracker = decode_from_json(TRANSACTIONS_JSON_PATH, BudgetTracker)
```

```python
budge_tracker.balance = 1000
budge_tracker.balance
```

```python
for cat in cats:
    budge_tracker.add_category(cat)
```

```python
transactions = [
    Transaction('На шторы', 499, TransactionType.expense, 'школа'),
    Transaction('ЗП', 500, TransactionType.income, 'работа'),
    Transaction('Путевка в Крым', 10000, TransactionType.expense, 'отдых'),
    Transaction('Еще на шторы', 2, TransactionType.expense, 'школа')
]
```

```python
for transaction in transactions:
    if (budge_tracker.is_over_limit(transaction)):
        print(f'{transaction} добавлена с ПРЕВЫШЕНИЕМ ЛИМИТА')
    else:
        print(f'{transaction} добавлена')
    budge_tracker.add_transaction(transaction)
```

```python
print(budge_tracker.str_sorted_by_category())
```

```python
print(budge_tracker.str_sorted_by_category(income=False))
```

```python
print(budge_tracker.str_sorted_by_category(expense=False))
```
