#### Упражнение №5

Сгенерировать, используя модуль псевдослучайных чисел random , или ввести с
клавиатуры список целых чисел. Вывести в консоль, затем перевернуть его и снова
вывести в консоль.

```python
def get_random_int_list(a: int = None, size: int = 10, start: int = 10, end: int = 99) -> list:
    random.seed(a=a)
    return [random.randint(start, end) for _ in range(size)]
```

```python
random.seed(a=0)
random_list = get_random_int_list(a=0)
print(f'Исходный список: {random_list}')
random_list.reverse()
print(f'Перевернутый список: {random_list}')
```

#### Упражнение №6

Сгенерировать, используя модуль псевдослучайных чисел random , или ввести с
клавиатуры два списка целых чисел. Вывести их в консоль. Создать новый пустой
список. Добавить в него все четные (по индексу) элементы первого списка и все
нечетные (по индексу) элементы второго списка. Вывести третий список в консоль.

```python
def get_even_odd_idx_list(l1: list, l2: list) -> list:
    result_list = []
    for i in range(0, len(l1), 2):
        result_list.append(l1[i])

    for i in range(1, len(l2), 2):
        result_list.append(l2[i])
    return result_list
```

```python
random_list_1 = get_random_int_list(a=52)
random_list_2 = get_random_int_list(a=1)
print(f'Исходный список 1: {random_list_1}\nИсходный список 2: {random_list_2}')
print(f'Итоговый список 3: {get_even_odd_idx_list(random_list_1, random_list_2)}')
```

#### Упражнение №7

Сгенерировать, используя модуль псевдослучайных чисел random , или ввести с
клавиатуры список произвольных элементов (целые числа, числа с плавающей
точкой, строки). Вывести в консоль. Убрать из него все дубликаты через
приведение типов. Вывести в консоль.


```python
def get_random_type_list(a: int = None, size: int = 10) -> list:
    random.seed(a=a)
    result_list = []
    types = [int, float, str]
    for _ in range(size):
        choice = random.choice(types)
        if choice is int:
            result_list.append(random.randint(0, 3))
        elif choice is float:
            result_list.append(random.uniform(0, 3))
        elif choice is str:
            result_list.append(''.join(random.choices('abc', k=1)))
    return result_list
```

```python
orig_list = get_random_type_list(a=20)
unique_list = list(dict.fromkeys(orig_list))

print(f'Исходный список: {orig_list}')
print(f'Итоговый список: {unique_list}')
```

#### Упражнение №8

Сгенерировать, используя модуль псевдослучайных чисел random , или ввести с
клавиатуры словарь, где ключом является строка, значением — целое число или
число с плавающей точкой. Вывести в консоль. Для всех уникальных значений
создать кортеж, где первым элементом будет значение, вторым — список
связанных с ним ключей. Собрать эти кортежи в список, вывести его в консоль.


```python
def get_random_dict(a: int = None, size: int = 10, start_int: int = 10, end_int: int = 99) -> dict:
    random.seed(a=a)
    letters = 'abcdef'
    return {''.join(random.choices(letters, k=1)): random.randint(start_int, end_int) for _ in range(size)}
```

```python
def get_tuples_from_dict(d: dict) -> list:
    set_items = set(d.values()) # set()
    result_list = []
    for item in set_items:
        temp_list = [item]
        for key, value in d.items():
            if item == value:
                temp_list.append(key)
        temp_tuple = tuple(temp_list)
        result_list.append(temp_tuple)
    return result_list
```

```python
random_dict = get_random_dict(a=5, start_int=0, end_int=3)
tuple_list = get_tuples_from_dict(random_dict)

print(f'Исходный словарь: {random_dict}')
print(f'Итоговый список кортежей: {tuple_list}')
```

#### Упражнение №9

Сгенерировать, используя модуль псевдослучайных чисел random , или ввести с
клавиатуры два словаря, где ключом является строка, значением — целое число
или число с плавающей точкой. Вывести в консоль. Найти пересечения множеств
значений словарей. Создать новый словарь, содержащий только те пары ключзначение, значения из которых входит в пересечение. Вывести в консоль.

```python
def get_difference(d1: dict, d2: dict) -> tuple[set, dict]:
    set_values_1 = set(d1.values())
    set_values_2 = set(d2.values())
    diff_items = set_values_1.intersection(set_values_2)
    result_dict = {}
    for item in diff_items:
        for key, value in d1.items():
            result_dict.setdefault(key, value) if value == item else None
        for key, value in d2.items():
            result_dict.setdefault(key, value) if value == item else None
    return diff_items, result_dict
```

```python
dict_1 = get_random_dict(a=2, size=5, start_int=0, end_int=5)
dict_2 = get_random_dict(a=3, size=10, start_int=0, end_int=5)

diff_items, diff_dict = get_difference(dict_1, dict_2)

print(f'Исходный словарь 1: {dict_1}\nИсходный словарь 2: {dict_2}')
print(f'Пересечение множеств значений: {diff_items}')
print(f'Итоговый словарь: {diff_dict}')
```
